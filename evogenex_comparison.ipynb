### Comparison with EvoGeneX on public cell line data
[Hirsch et al (2025)](https://www.sciencedirect.com/science/article/pii/S2405471224003491) used their methods EvoGeneX on their melanoma cell line data to detect genes which change due to selection versus neutral evolution. They also estabilished a simulation framework to test their method on. We follow their approach to assess the performance of scPhyTr in classifying genes.


## Data

## EvoGeneX
Fits each gene separately and has no noise model.

## scPhyTr
Fits all genes jointly and uses a noise model. It may also use covariates in order to not confound adaptation for a condition-dependent variance.
